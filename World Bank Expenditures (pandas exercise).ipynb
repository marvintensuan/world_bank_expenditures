```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('World_Bank_Expenditures_by_Organizational_Unit.csv')
```

```python
df
```

From `df`, we can see interesting observations:

- An Organizational Unit such as 'Operational units' has a corresponding 'Operational units Total' which has its value on Subtotals instead of Actuals.
- However, from the shown data, Actuals and Subtotals don't seem to add up.

```python
df = df.rename(columns = {'Actuals (US$, Millions)': 'Actuals', 'Subtotals (US$, Millions)': 'Subtotals',
                          'Vice Presidential Unit or Grouping': 'Grouping'})
```

```python
df
```

```python
pivot_table_1 = pd.pivot_table(df, values = 'Actuals', index = ['Organizational Unit', "Grouping"],
                               aggfunc = np.sum, fill_value = 0)
```

```python
pivot_table_1
```

It looks like after all, only 'Operational units' has a Grouping which has an 'Operational units Total'.

```python
5552.2+2028.2 #Total Actuals from Operational Units.
```

```python
pivot_table_2 = pd.pivot_table(df, values = 'Subtotals', index = ['Grouping'],
                               aggfunc = np.sum, fill_value = 0)
```

```python
pivot_table_2
```

Still, the actuals we computed from Operational units Actuals does not equal to the subtotals column.

```python
df_operational_units = df[df['Organizational Unit'] == 'Operational units']
```

```python
df_operational_units
```

Now, we can see that the reason why the Actuals and Subtotals for 'Operational Units' is because it does not have Actuals data prior to 2015.
