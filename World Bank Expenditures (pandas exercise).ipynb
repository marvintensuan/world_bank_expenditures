```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('World_Bank_Expenditures_by_Organizational_Unit.csv')
```

```python
df
```

Below we will see a table summarizing the overall expenditures per Organizational Unit.

```python
pivot_table_1 = pd.pivot_table(df, values = 'Actuals (US$, Millions)', index = ['Organizational Unit'],
                               aggfunc = np.sum, fill_value = 0)
```

```python
pivot_table_1
```

On the next table, we try to see the Actuals data per year.

```python
pivot_table_2 = pd.pivot_table(df, values = 'Actuals (US$, Millions)', index = ['Organizational Unit'],
                               columns = ['Fiscal Year'],aggfunc = np.sum, fill_value = 0)
```

```python
pivot_table_2
```
